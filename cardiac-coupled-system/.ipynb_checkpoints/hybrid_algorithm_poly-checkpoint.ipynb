## Direct Linear Regression


```python
def compute_linear_weight(x,y):
    y = np.reshape(y,(-1,1))
    A = np.matmul(np.transpose(x),x)
    b = np.matmul(np.transpose(x),y)
    w = np.matmul(np.linalg.inv(A),b)
    y_pred = np.matmul(x,w)
    return w, y_pred
```

```python
fusion_data = np.concatenate((np.reshape(x_data_scaler[:,0],(-1,1)),lace_data_scaler),axis = 1)
```

## Build Polynomial Dictionray

```python
class DicPoly(object):
    def __init__(self, degree = 2):
        self.dictionary_degree = degree
    
    def call(self, data):
        poly = []
        poly_index = []
        pre_poly = []
        N = np.shape(data)[1]
        for j in range(N):
            pre_poly.append([j])
        for i in range(1, self.dictionary_degree):
            cur_poly = []
            for pp in pre_poly:
                index = pp[-1]
#                 print(index)
                for j in range(index, N):
                    cur_term = pp[:]
                    cur_term.append(j)
                    cur_poly.append(cur_term)
                    poly_index.append(cur_term)
            pre_poly = cur_poly
#             print(i)
        for index in poly_index:
            cur_poly_data = tf.cast(tf.ones(tf.shape(data)[0]),dtype = 'float64')
            for i in index:
                cur_poly_data = cur_poly_data * data[:,i]
            poly.append(cur_poly_data)
        poly = tf.transpose(poly)
        poly = tf.convert_to_tensor(poly)
        ones = tf.cast(tf.ones((tf.shape(data)[0],1)),dtype = 'float64')
        return tf.concat((ones,data, poly),axis=1)
```

```python
# Neural Network for K(u)
from tensorflow.keras.layers import Layer, Dense
from tensorflow.keras.layers import Input, Add, Multiply, Lambda, Concatenate
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers.legacy import Adam
import numpy as np
import tensorflow as tf
tf.keras.backend.set_floatx('float64')

class KNN(Layer):
    def __init__(self, n_K=27, layer_sizes=[32, 32], **kwargs):
        self.layer_sizes = layer_sizes
        self.n_K = n_K
        self.input_layer = Dense(self.layer_sizes[0], name='K_input')
        self.hidden_layers = [Dense(layer_sizes[i], activation='tanh', name='K_hidden_%d'%i) for i in range(len(layer_sizes))]
        self.output_layer = Dense(self.n_K**2, name='K_output')
        
    def call(self, inputs_u):
        K = self.input_layer(inputs_u)
        for layer in self.hidden_layers:
            K = K + layer(K)
#         outputs = K
        outputs = self.output_layer(K)
#         return outputs
        return tf.reshape(outputs,(-1,self.n_K,self.n_K))
```

```python
dic = DicPoly(2)
knn = KNN(15)
```

```python
target_dim = 4
u_dim = 1
```

```python
# Build Model
inputs_x = Input((target_dim,))
inputs_y = Input((target_dim,))
inputs_u = Input((u_dim,))
        
psi_x = dic.call(inputs_x)
psi_y = dic.call(inputs_y)

model_k = Model(inputs=inputs_u,outputs=knn.call(inputs_u))
# outputs = tf.matmul(psi_x,model_k(inputs_u)) - psi_y
psi_x = tf.expand_dims(psi_x, 1)
psi_y = tf.expand_dims(psi_y, 1)
outputs = tf.matmul(psi_x,model_k(inputs_u))-psi_y
outputs = tf.reshape(outputs,(tf.shape(psi_x)[0],-1))
model_KoopmanDL = Model(inputs=[inputs_x, inputs_y, inputs_u], outputs=outputs)
```

```python
# Koopman with polynomial 
lr = 0.001
opt = tf.keras.optimizers.Adam(learning_rate=lr)
model_KoopmanDL.compile(optimizer=opt, loss='mse')
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test, u_train, u_test = train_test_split(x_data_scaler, y_data_scaler, u_data_scaler, test_size=0.33, random_state=None)
```

```python
iters = 2
epochs = 30
zeros_data_y_train = tf.zeros_like(dic.call(y_train))
zeros_data_y_test = tf.zeros_like(dic.call(y_test))
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
  try:
    tf.config.experimental.set_virtual_device_configuration(
        gpus[0],[tf.config.experimental.VirtualDeviceConfiguration(memory_limit=5120)])
  except RuntimeError as e:
    print(e)
history = model_KoopmanDL.fit([x_train, y_train, u_train], zeros_data_y_train, validation_data = ([x_test, y_test, u_test], zeros_data_y_test),epochs=epochs)
```

```python
def predict_v_koopman(dic_psi, model_k, x_train, u_train):
    psi_x = dic_psi.call(x_train)
    psi_x = tf.expand_dims(psi_x, 1)
    outputs = tf.matmul(psi_x,model_k(u_train))
    outputs = tf.reshape(outputs,(tf.shape(psi_x)[0],-1))
    return outputs[:,1:5]
```

```python
y_direct_koopman = predict_v_koopman(dic, model_k, x_data_scaler, u_data_scaler)
```
