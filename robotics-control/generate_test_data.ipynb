```python
dlt_t = 1
h = 5
tt = np.arange(0, 180, dlt_t)
N = len(tt)
N
```

```python
# alpha_0_test = 1/6 * np.pi * np.sin(np.pi / 90 * tt - np.pi/2)
alpha_0_test = 1/6 * np.pi * np.ones(N)

alpha_a_test = 1/18 * np.pi * np.ones(N)
omega_a_ref = 2 * np.pi * np.ones(N)
omega_a_test = 2 * np.pi + 1 * np.pi * np.sin(np.pi/60 * tt)
omega_a_test3 = 2 * np.pi + 1 * np.pi * np.sin(np.pi/10 * tt)
```

```python
import matplotlib.pyplot as plt
plt.plot(omega_a_test3)
```

```python
# 19; 31; 8; 3; 97; 62; 77; 11; 9; 6
seeds = [19, 31, 8, 3, 97, 62, 77, 11, 9, 6]
suffix = 8
np.random.seed(seeds[suffix])
x = np.random.uniform(-0.5, 0.5, 1)
z = np.random.uniform(-0.1, 0.1, 1)
psi = np.random.uniform(-np.pi/4, np.pi/4, 1)
v_x = np.random.uniform(0, 0.04, 1)
v_y = np.random.uniform(-0.0025, 0.0025, 1)
omega = np.random.uniform(-0.5, 0.5, 1)
omega_a_test2 = np.random.uniform(0, 10 * np.pi, N)
s0 = np.column_stack((x, z, psi, v_x, v_y, omega))
s0 = s0.flatten()
s0
```

```python
dlt_t
```

```python
t_span = (0, dlt_t)
t_eval = [0, dlt_t]
t_eval
```

```python
# Build traj
s_des_test = [s0]
s1 = s0
for i in range(N):
    s1 = dynamics_solver(s1, [alpha_0_test[i], alpha_a_test[i]], omega_a_ref[i:i+1])
    s_des_test.append(s1)
```

```python
s_des_test = np.row_stack(s_des_test)
```
